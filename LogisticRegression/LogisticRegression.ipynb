```python
from LogisticRegression import LogisticRegression
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 训练数据
X1 = np.random.randn(100,2) + np.array([2,3])
X2 = np.random.randn(100,2) + np.array([5,8])
Y1 = np.zeros(100)
Y2 = np.ones(100)
X = np.concatenate((X1,X2),axis=0)
Y = np.append(Y1,Y2)
```

```python
# 测试数据
X3 = np.random.randn(100,2) + np.array([2,3])
X4 = np.random.randn(100,2) + np.array([5,8])
Y3 = np.zeros(100)
Y4 = np.ones(100)
test_X = np.concatenate((X3,X4),axis=0)
test_Y = np.append(Y3,Y4)
```

```python
model = LogisticRegression()
```

```python
model.train(X,Y)
```

```python
x1 = np.linspace(0, 10, 50)
w1 = model.w[0]
w2 = model.w[1]
x2 = -(w1 * x1 + model.b)/w2

plt.figure()
plt.scatter(X1[:,0],X1[:,1])
plt.scatter(X2[:,0],X2[:,1])
plt.plot(x1,x2,color='red', linestyle='-')
plt.xlim(0,10)
plt.ylim(0,10)
```

```python
# 测试集分数
model._score(test_X, test_Y)
```
